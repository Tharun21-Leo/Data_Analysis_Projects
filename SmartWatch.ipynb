```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px


watch_data = pd.read_csv("unclean_smartwatch_health_data.csv")

watch_data
```

```python
watch_data.isnull().sum()
```

```python
watch_data["User ID"].dropna(inplace = True)
```

```python
watch_data["Sleep Duration (hours)"] = pd.to_numeric(watch_data["Sleep Duration (hours)"], errors="coerce")
```

```python
watch_data["Stress Level"] = pd.to_numeric(watch_data["Stress Level"] , errors=  "coerce")
watch_data
```

```python
percent_null = watch_data.isnull().mean() * 100
print(percent_null)
```

```python
watch_data.dropna(inplace=True)
```

```python
watch_data["User ID"].mean()
watch_data["Heart Rate (BPM)"].mean()
watch_data['Blood Oxygen Level (%)'].mean()
watch_data['Step Count'].mean()
watch_data['Sleep Duration (hours)'].mean()
watch_data['Activity Level'].mode()
watch_data['Stress Level'].mean()
```

```python
watch_data
```

```python
watch_data['Activity Level'].unique()
```

```python
filtered_data = watch_data[(watch_data["Sleep Duration (hours)"] > 6)& (watch_data['Activity Level'] == 'Highly Active')]
filtered_data
```

```python
#Average Steps per User
avg_steps = watch_data.groupby("User ID")["Step Count"].mean()
print(avg_steps)
```

```python
#Percentage of users meet the 10,000 daily steps benchmark

users_meeting_10k = (avg_steps >= 10000).sum()
total_users = avg_steps.shape[0]
pct_meeting_10k = (users_meeting_10k / total_users) * 100
print("Percentage of users meeting 10k steps:", pct_meeting_10k)
```

```python
#correlation between steps and Blood Oxygen Level?
correlation = watch_data["Step Count"].corr(watch_data["Blood Oxygen Level (%)"])
print("Correlation between Steps and Blood Oxygen Level:", correlation)
```

```python
corr_matrix = watch_data[["Step Count", "Blood Oxygen Level (%)"]].corr()
print(corr_matrix)
```

```python
sns.lmplot(
    x="Step Count", 
    y="Blood Oxygen Level (%)", 
    data=watch_data, 
    line_kws={"color":"red"}, 
    scatter_kws={"alpha":0.5}
)

corr_value = watch_data["Step Count"].corr(watch_data["Blood Oxygen Level (%)"])

# Add correlation text on the plot
plt.text(
    x=watch_data["Step Count"].min(), 
    y=watch_data["Blood Oxygen Level (%)"].max(), 
    s=f"Pearson Corr = {corr_value:.2f}", 
    fontsize=12, color="blue"
)

plt.show()
```
